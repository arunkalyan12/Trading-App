```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv1D, LSTM, Dropout, Layer, Dense
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras import backend as K
import tensorflow as tf
```

```python
#Read data from "preprocessing.csv"
data = pd.read_csv("..\Data\Preprocessed\Preprocessed.csv")

X_features = data.drop(columns=['label', 'label_prob'])
y = data['label'].values

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X_features)

def create_sequences(features, targets, seq_length, stride=5):
    X_seq, y_seq = [], []
    for i in range(0, len(features) - seq_length, stride):
        X_seq.append(features[i:i+seq_length])
        y_seq.append(targets[i+seq_length])
    return np.array(X_seq, dtype=np.float32), np.array(y_seq, dtype=np.float32)

SEQ_LENGTH = 60
X, y = create_sequences(X_scaled, y, SEQ_LENGTH, stride=5)

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.1, shuffle=False)

y_train = y_train + 1
y_val = y_val + 1
```
